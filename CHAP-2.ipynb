```python
#2 (if statement is false)

light="green"
if(light=="red"):
    print("STOP")
elif(light=="green"):
    print("GO")
else:
    print("LOOK")
```

```python
#4 grade

marks=int(input("Enter Your Mark"))
if(marks >= 90):
    grade='A'
elif(marks>=80 and marks<90 ):
    grade='B'
elif(marks>=70 and marks<80 ):
    grade='C'
elif(marks>=60 and marks<70 ):
    grade='D'
elif(marks<60):
    grade='FAIL'
else:
    print("Enter Valid Mark")

print(grade)
print("Thank You")
    
```

```python
#5 (nesting if else)

age=75
if(age>=18):
    if(age>=80):
        print("You can't drive")
    else:
        ("You can drive")
else:
    print("you can't drive")
```
