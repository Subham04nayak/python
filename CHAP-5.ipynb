```python
#4- print the elements of list [1,4,9,16,25,36,49,64,81,100]

list=[1,4,9,16,25,36,49,64,81,100]
idx=0
while idx<len(list):
    print(list[idx])
    idx +=1
```

```python
#5- search for the element in the tuple (1,4,9,16,25,36,49,64,81,100)

nums=(1,4,9,16,25,36,49,64,81,100)
x=49
idx=0
while idx<=len(nums):
    if(nums[idx]==x):
        print("FOUND at " , idx)
    idx+=1
    
```

### while with else

```python
i=10
while(i>=1):
    print(i)
    i-=1
else:
    print("stop")
```

```python
i=-10
while(i>=1):
    print(i)
    i-=1
else:
    print("stop")   #since the while statement is false so it prints else part
```
