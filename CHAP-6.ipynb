```python
def top_three_digits(num):
    digits = sorted(set(str(num)), reverse=True)
    return digits[0], digits[1], digits[2]

print(top_three_digits(6328))
```

```python
def divisible_numbers():
    for i in range(100, 501):
        if i % 3 == 0 and i % 5 == 0:
            print(i,end=",")

divisible_numbers()
```

```python
def divisible_numbers():
    return [i for i in range(100, 501) if i % 3 == 0 and i % 5 == 0]

print(divisible_numbers())
```

```python
def vowel_or_consonant(ch):
    return "Vowel" if ch.lower() in "aeiou" else "Consonant"
vowel_or_consonant("i")
```

```python
def check_vowel_or_consonant(ch):
    if ch.isalpha() and len(ch) == 1:
        if ch.lower() in ['a', 'e', 'i', 'o', 'u']:
            return "Vowel"
        else:
            return "Consonant"
    else:
        return "Invalid input"
check_vowel_or_consonant("aertt")
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a+b
    return a
fibonacci(5)
```
