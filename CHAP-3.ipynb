```python
lst = []

while True:
    print("1.Create  2.Display  3.Insert  4.Delete  5.Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        n = int(input("Enter N: "))
        lst = [int(input()) for _ in range(n)]

    elif choice == 2:
        print(lst)

    elif choice == 3:
        pos = int(input("Position: "))
        val = int(input("Value: "))
        lst.insert(pos, val)

    elif choice == 4:
        pos = int(input("Position: "))
        lst.pop(pos)

    elif choice == 5:
        break
```

```python
def remove_all(list, x):
    return [i for i in list if i != x]
remove_all([12,34,56,34,32],34 )
```

```python
def multiples_list(n):
    return [[i*j for j in range(1, 6)] for i in range(1, n+1)]
multiples_list(3)
```

```python
def print_matrix(mat):
    for row in mat:
        for val in row:
            print(val, end="\t")
        print()
print_matrix(1)
```

```python
def square_list(lst):
    return [i*i for i in lst]
square_list([12,34,56,76])
```
