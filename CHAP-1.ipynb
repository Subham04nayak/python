# OPERATORS
### 1-ARITHMETIC

```python
a=10
b=5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a//b) #floor division(2.9->2, 2.0->2)
print(a%b) #modulo oprator-remainder
print(a**b) #power operator
```

### 2-RELATIONAL

```python
a=50
b=20
print(a==b)
print(a!=b)
print(a>b)
print(a>=b)
print(a<=b)
print(a<b)
```

### 3-ASSIGNMENT 

```python
num=7
num1=7
num2=7
num3=7
num4=7
num5=7

num=num+5 # num+=5
num1 -=5
num2 *=5
num3 /=5
num4 %=5
num5 **=5

print(num)
print(num1)
print(num2)
print(num3)
print(num4)
print(num5)
```

### 4-LOGICAL

```python
val1=True
val2=False

print(not val1)
print(val1 and val2)
print(val1 or val2)
print(val1 and val1)
print(val2 or val2)
```

```python
a=100
b=30
print(not(b>a))
print(a==b and True)
print(a>b or b>a)
```
