### nested dict.

```python
student={
    "name" : "Subham",
    "subj" : {
            "phy":45,
            "chem":34,
            "math":23
        } ,
    "age" : 21
}

print(student)
```

```python
print(student["subj"])
```

```python
print(student["subj"]["math"])
```

### Methods

```python
#keys()

info1={
    "name":"Subham Nayak",
    "age":21,
    "marks":[23,21,5,15],
    "weight":73.5,
    "male":True
}

print(info1.keys())
```

```python
print(list(info1.keys()))  # converted to list
```

```python
print(len(list(info1.keys())))
```

```python
# values()

print(info1.values())
```

```python
#items

print(info1.items())
```

```python
print(list(info1.items()))  #converted to list
```

```python
pairs=list(info1.items())
print(pairs[0])
```

```python
# get("key")

info1={
    "name":"Subham Nayak",
    "age":21,
    "marks":[23,21,5,15],
    "weight":73.5,
    "male":True
}
print(info1.get("marks"))
```

```python
# update(new_dict)

info1={
    "name":"Subham Nayak",
    "age":21,
    "marks":[23,21,5,15],
    "weight":73.5,
    "male":True
}
new_info = {"name" : "kiran nayak",
            "age" : 20
           }
info1.update(new_info)
print(info1)
```

```python

cubes = {x: x**3 for x in range(1, 6)}
print(cubes)
```

```python
nums = [10, 3, 7, 1, 9, 4, 2, 8, 5, 6]
squares = (x**2 for x in nums if x % 2 != 0)

print(squares)
```

```python
import numpy as np

# Arrays
a = np.ones((2, 3))
b = np.zeros((3, 3))
c = np.full((2, 5),7)

print(a)
print(b)
print(c)

# 2x2 array using arange
arr = np.arange(4).reshape(2, 2)
print(arr)

# (a) Cube every element
print(arr ** 3)

# (b) Add 7 to every element
print(arr + 7)

# (c) Multiply every element by 2
print(arr * 2)
```
